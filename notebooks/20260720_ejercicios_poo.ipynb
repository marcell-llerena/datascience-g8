```python
# Ejercicio 2
class Rectangulo:
    def __init__(self, base: float, altura: float) -> None:
        self.base = base
        self.altura = altura

    def area(self) -> float:
        return self.base * self.altura

    def perimetro(self) -> float:
        return 2 * (self.base + self.altura)

    def __str__(self) -> str:
        return f"Rectangulo de base: {self.base} y altura: {self.altura}"

mi_rectangulo = Rectangulo(3, 5)
print(mi_rectangulo.area())
print(mi_rectangulo.perimetro())
print(mi_rectangulo)
```

```python
class Animal:
    def __init__(self, nombre: str, edad: int) -> None:
        self.nombre = nombre
        self.edad = edad

    def hacer_sonido(self):
        print("Grrrrr")


class Perro(Animal):
    def __init__(self, nombre: str, edad: int) -> None:
        super().__init__(nombre, edad)

    def hacer_sonido(self):
        print("Gua gua!")

class Gato(Animal):
    def __init__(self, nombre: str, edad: int) -> None:
        super().__init__(nombre, edad)

    def hacer_sonido(self):
        print("Miau Miau!")


perro_1 = Perro("Hachiko", 3)
perro_2 = Perro("Fido", 5)
gato_1 = Gato("Lucifer", 10)
vaca_1 = Animal("Lola", 14)

animales = [perro_1, perro_2, gato_1, vaca_1]

for animal in animales:
    animal.hacer_sonido()
```

```python
# Ejercicio 8
class Reproductor:
    def __init__(self):
        self.playlist = []
        self.posicion = 0

    def agregar_cancion(self, cancion: str) -> None:
        self.playlist.append(cancion)

    def cancion_actual(self) -> str:
        if not self.playlist:
            return "No hay canciones en la playlist"

        return self.playlist[self.posicion]

    def siguiente(self) -> str:
        self.posicion += 1
        if self.posicion < len(self.playlist):
            return self.cancion_actual()

        self.posicion -= 1
        return self.cancion_actual()

    def anterior(self) -> str:
        self.posicion -= 1
        if self.posicion >= 0:
            return self.cancion_actual()

        self.posicion += 1
        return self.cancion_actual()


mi_reproductor = Reproductor()
mi_reproductor.agregar_cancion("The trooper")
mi_reproductor.agregar_cancion("Dance of death")
mi_reproductor.cancion_actual()
mi_reproductor.siguiente()
mi_reproductor.cancion_actual()
```

```python
# Ejercicio 9
class Motor:
    def __init__(self, tipo: str, caballos: float) -> None:
        self.tipo = tipo
        self.caballos = caballos
        self.encendido = False

    def arrancar(self) -> str:
        if self.encendido:
            return "El motor ya esta encendido"

        self.encendido = True
        return f"El motor {self.tipo} de {self.caballos} HP se encendio"


class Vehiculo:
    def __init__(self, marca: str, motor: Motor) -> None:
        self.marca = marca
        self.motor = motor

    def arrancar(self) -> str:
        return self.motor.arrancar()

motor = Motor("gasolina", 150)
vehiculo = Vehiculo("Ford", motor)

print(vehiculo.arrancar())
print(vehiculo.arrancar())
```

```python
# Ejercicio 10

# Crea tres clases que trabajen juntas: Libro, Usuario y Biblioteca.
# Libro: titulo, autor y disponible (True por defecto).
# Usuario: nombre y una lista de libros_prestados.
# Biblioteca: guarda una lista de libros y ofrece los métodos agregar_libro(), buscar(titulo), prestar(titulo,
# usuario), devolver(titulo, usuario) y catalogo().
# Al prestar un libro debe marcarse como no disponible y añadirse a la lista del usuario; al devolverlo, lo
# contrario.
# Contempla los casos: el libro no existe, el libro ya está prestado, el usuario no tiene ese libro.

class Libro:
    def __init__(self, titulo: str, autor: str) -> None:
        self.titulo = titulo
        self.autor = autor
        self.disponible = True


class Usuario:
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre
        self.libros_prestados = []

class Biblioteca:
    def __init__(self):
        self.libros = []

    def agregar_libro(self, libro: Libro) -> None:
        self.libros.append(libro)

    def buscar_libro(self, titulo: str) -> Libro | None:
        for libro in self.libros:
            if libro.titulo.lower() == titulo.lower():
                return libro

    def prestar_libro(self, titulo: str, usuario: Usuario) ->  str:
        libro = self.buscar_libro(titulo)

        if libro is None:
            return "El libro no se encuentra en la biblioteca"

        if not libro.disponible:
            return "El libro ya fue prestado"

        libro.disponible = False
        usuario.libros_prestados.append(libro)

        return f"Libro prestado: {libro.titulo} al usuario: {usuario.nombre}"

    def devolver(self, titulo: str, usuario: Usuario) -> str:
        libro = self.buscar_libro(titulo)

        if libro is None:
            return "No existe el libro"

        if libro not in usuario.libros_prestados:
            return f"{usuario.nombre} no tiene prestado el libro: {titulo}"

        usuario.libros_prestados.remove(libro)
        libro.disponible = True

        return f"Libro: {titulo} fue devuelto por el usuario: {usuario.nombre}"

    def catalogo(self) -> None:
        for libro in self.libros:
            print(f"Titulo: {libro.titulo} | Autor: {libro.autor} | Estado: {libro.disponible}")




biblioteca = Biblioteca()
libro_1 = Libro("Mi planta de naranja lima", "Jose Mauro de Vasconcelos")
libro_2 = Libro("Quien mato a Palomino Molero", "Mario Vargas Llosa")

biblioteca.agregar_libro(libro_1)
biblioteca.agregar_libro(libro_2)

biblioteca.catalogo()

usuario_1 = Usuario("Pedro")
usuario_2 = Usuario("Alejandra")

print("="*20)

print (
    biblioteca.prestar_libro(
        "Quien mato a Palomino Molero",
        usuario_1
    )
)
print("="*20)

biblioteca.catalogo()

print("="*20)
for libro in usuario_1.libros_prestados:
    print(libro.titulo)

print("="*20)
print(biblioteca.devolver(
    "Quien mato a Palomino Molero",
    usuario_2
))

# print(biblioteca.devolver(
#     "Quien mato a Palomino Molero",
#     usuario_1
# ))

print("="*20)
for libro in usuario_1.libros_prestados:
    print(libro.titulo)

print("="*20)
print (
    biblioteca.prestar_libro(
        "La Metamorfosis",
        usuario_1
    )
)
print("="*20)

print (
    biblioteca.prestar_libro(
        "Quien mato a Palomino Molero",
        usuario_2
    )
)
```
