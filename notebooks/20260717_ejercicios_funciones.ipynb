```python
def analizar_ventas(ventas: list[int | float]) -> dict[str, int | float]:
    if not ventas:
        return "No hay ventas"

    for venta in ventas:
        if venta < 0:
            return f"Hay ventas negativas: {venta}"

    total_vendido = sum(ventas)
    promedio = total_vendido / len(ventas)
    venta_mas_alta = max(ventas)
    venta_mas_baja = min(ventas)

    total_superior_promedio = 0
    total_ventas_inferior_100 = 0

    for venta in ventas:
        if venta > promedio:
            total_superior_promedio = total_superior_promedio + 1

        if venta < 100:
            total_ventas_inferior_100 = total_ventas_inferior_100 + 1

    return {
        "total_vendido": total_vendido,
        "promedio": promedio,
        "venta_mas_alta": venta_mas_alta,
        "venta_mas_baja": venta_mas_baja,
        "total de ventas superior al promedio": total_superior_promedio,
        "total de ventas inferior a 100": total_ventas_inferior_100
    }


analizar_ventas(ventas)
```

```python
# Ejercicio 2
def validar_password(password: str) -> tuple[bool, list[str]]:

    simbolos_requeridos = ("@", "#", "$",  "%")

    errores = []

    if len(password) < 8:
        errores.append(
            "El password tiene menos de 8 caracteres"
        )

    contiene_minuscula = False
    contiene_mayuscula = False
    contiene_numero = False
    contiene_simbolos_requeridos = False

    for caracter in password:
        if caracter.islower():
            contiene_minuscula = True
            break

    for caracter in password:
        if caracter.isupper():
            contiene_mayuscula = True
            break

    for caracter in password:
        if caracter.isnumeric():
            contiene_numero = True
            break

    for caracter in password:
        if caracter in simbolos_requeridos:
            contiene_simbolos_requeridos = True
            break

    if not contiene_minuscula:
        errores.append(
            "El password no contiene letra minuscula"
        )

    if not contiene_mayuscula:
        errores.append(
            "El password no contiene letra mayuscula"
        )

    if not contiene_numero:
        errores.append(
            "El password no contiene caracter numerico"
        )

    if not contiene_simbolos_requeridos:
        errores.append(
            "El password no contiene simbolos requeridos"
        )

    if len(password) != len(password.strip()):
        errores.append(
            "El password contiene espacios en blanco"
        )

    es_valida = False if errores else True

    return (es_valida, errores)
```

```python
es_valida, errores = validar_password('python@')
print(es_valida)
print(errores)
```

```python
# Ejercicio 3

def registrar_estudiante(
    estudiantes: dict[str, int | float | str],
    nombre: str,
    notas: list[int | float]
) -> str:
    if not nombre:
        return "Nombre invalido"

    if nombre in estudiantes:
        return "El nombre ya existe, no se admite duplicados"

    for nota in notas:
        if isinstance(nota, str):
            return "Hay notas no numericas"
        else:
            if nota < 0 or nota > 20:
                return "Hay notas con valores menores a 0 o mayores a 20"

    promedio = round(sum(notas) / len(notas), 2)

    if promedio >= 11:
        estado = "Aprobado"
    else:
        estado = "Desaprobado"

    estudiantes[nombre] = {
        "estado": estado,
        "promedio": promedio,
        "notas": notas
    }

    return "El registro fue exitoso"
```

```python
estudiantes = {}

registrar_estudiante(
    estudiantes,
    'Ana',
    [18, 15, 17]
)
```

```python
# Ejercicio 4
palabras = [
    'python', 'Java', 'código', 'web',
    'java', 'datos', 'API', 'lista', 10000
]
# resultado = agrupar_por_longitud(palabras)

def agrupar_por_longitud(palabras: list[str]) -> dict[int, list[str]]:
    palabras_limpias = set([
        palabra.lower()
        for palabra in palabras
        if isinstance(palabra, str)
    ])

    print(palabras_limpias)

    grupos = {}
    for palabra_limpia in palabras_limpias:
        longitud = len(palabra_limpia)
        grupos.setdefault(longitud, []).append(palabra_limpia)

    print(grupos)

    for llave in grupos:
        grupos[llave].sort()

    print(grupos)

    return grupos

print( agrupar_por_longitud(palabras) )
```

```python
# Ejercicio 5


# actualizar_inventario(
#     inventario, 'Mouse', 4, 'salida'
# )

# actualizar_inventario(inventario, producto, cantidad, operacion)

def actualizar_inventario(
    inventario: dict[str, int],
    producto: str,
    cantidad: int,
    operacion: str  #ACTUALIZACION
) -> dict[str, str | int]:
    operaciones = ("entrada", "salida")
    operacion = operacion.lower() #entrada #actualizacion

    if operacion not in operaciones:
        return {
            "mensaje": f"Error, tipo de operacion no soportada: {operacion}",
            "stock_actual": f"{inventario[producto]}"
        }

    if not isinstance(cantidad, int):
        return {
            "mensaje": f"Error, tipo de dato no adecuado, esperaba un entero y consiguio un {type(cantidad)}",
            "stock_actual": f"{inventario[producto]}"
        }

    if cantidad <= 0:
        return {
            "mensaje": "Error, la cantidad es negativa o es cero",
            "stock_actual": f"{inventario[producto]}"
        }

    if operacion == "entrada":
        if producto not in inventario:
            inventario[producto] = cantidad
        else:
            inventario[producto] += cantidad

    if operacion == "salida" and producto in inventario and inventario[producto] >= cantidad:
        inventario[producto] = inventario[producto] - cantidad

    return {
        "mensaje": "Operacion exitosa",
        "stock_actual": f"{inventario[producto]}"
    }
```

```python
inventario = {
    'Laptop': 5,
    'Mouse': 12,
    'Teclado': 8
}
```

```python
resultado = actualizar_inventario(inventario, "Mouse", 4, "salida")
print(inventario)
print(resultado)
```

```python
# Ejercicio 8

# comprimido = comprimir_texto(texto)
# original = descomprimir_texto(comprimido)

def comprimir_texto(texto: str) -> str:
    resultado = []

    caracter_actual = texto[0] # -> "a"
    contador = 1

    for caracter in texto[1:]: # -> "a"
        if caracter == caracter_actual:
            contador += 1
        else:
            resultado.append(f"{caracter_actual}{contador}")
            caracter_actual = caracter
            contador = 1

    resultado.append(f"{caracter_actual}{contador}")

    resultado = "".join(resultado)

    return resultado

texto = 'aaabbccccdaa'
comprimir_texto(texto)
```

```python
def descomprimir_texto(texto: str) -> str:
    letras = texto[0::2]
    numeros = texto[1::2]
    resultado = []
    for letra, numero in zip(letras, numeros):
        resultado.append(
            letra * int(numero)
        )

    resultado = "".join(resultado)

    return resultado

texto_encriptado = 'a3b2c4d1a2'
descomprimir_texto(texto_encriptado)
```
