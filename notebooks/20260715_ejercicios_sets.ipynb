### OPERADOS LOGICOS
- UNION -> | (OR)
- INTERSECCION -> & (AND)
- DIFERENCIA -> -
- DIFERENCIA SIMETRICA -> ^

```python
# Ejercicio 1

estudiantes_python = {"Ana","Luis","María","Carlos","Sofía","Diego"}
estudiantes_javascript = {"María","Carlos","Pedro","Lucía","Diego","Valeria"}

print("Estudiantes en ambos cursos", estudiantes_python & estudiantes_javascript)
print("Estudiantes en ambos cursos", estudiantes_python.intersection(estudiantes_javascript))

print("Estudiantes solo en Python", estudiantes_python - estudiantes_javascript)
print("Estudiantes solo en Javascript", estudiantes_javascript - estudiantes_python)

print("Todos los estudiantes", estudiantes_python | estudiantes_javascript)
```

```python
# Ejercicio 2
almacen_arequipa = {
    "laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"
}

almacen_lima = {
    "laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"
}
```

```python
print("Los productos disponibles en ambos almacenes: ", almacen_arequipa & almacen_lima)
print("Productos disponibles solo en AQP: ", almacen_arequipa - almacen_lima)
print("Productos disponibles solo en Lima: ", almacen_lima - almacen_arequipa)
print("Todos los productos: ", almacen_arequipa | almacen_lima)
print("Tienen el mismo inventario", almacen_arequipa == almacen_lima)
```

```python
# Ejercicio 3
permisos_usuario = {
    "leer", "editar", "comentar", "descargar"
}

permisos_requeridos = {
    "leer", "editar", "eliminar", "publicar"
}
```

```python
if len(permisos_requeridos - permisos_usuario) == 0:
    print("Tiene todos los permisos requeridos")
else:
    print(f"No tiene todos los permisos, le faltan los permisos: {permisos_requeridos - permisos_usuario} ")
```

```python
print("Los permisos que le faltan al usuario: ", permisos_requeridos - permisos_usuario)
```

```python
print("Los permisos adicionales que tiene el usuario: ", permisos_usuario - permisos_requeridos)
```

```python
print("Permisos diferentes: ", permisos_usuario ^ permisos_requeridos)
print("Total de Permisos diferentes: ", len(permisos_usuario ^ permisos_requeridos))
```

```python
# Ejercicio 4
numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}
```

```python
print("Elementos comunes: ", numeros_a & numeros_b & numeros_c)
print("Elementos de A y B pero no de C: ", (numeros_a & numeros_b) - numeros_c)
print("Solo A: ", numeros_a - (numeros_b | numeros_c))
print("Numeros unicos: ", (numeros_a ^ numeros_b) | (numeros_b ^ numeros_c) | (numeros_a ^ numeros_c))
```

```python
# Ejercicio 5
habilidades_candidato_1 = {
    "Python", "Git", "SQL", "Docker", "React"
}

habilidades_candidato_2 = {
    "Python", "JavaScript", "React", "Node.js", "Git"
}

habilidades_requeridas = {
    "Python", "Git", "SQL", "Docker"
}
```

```python
print("Habilidades en comun: ", habilidades_candidato_1 & habilidades_candidato_2)
print("Solo cantidato 1: ", habilidades_candidato_1 - habilidades_candidato_2)
print("Solo cantidato 2: ", habilidades_candidato_2 - habilidades_candidato_1)
print("LE faltan al candidato 1: ", habilidades_requeridas - habilidades_candidato_1)
print("LE faltan al candidato 2: ", habilidades_requeridas - habilidades_candidato_2)
print("Todas las habilidades disponibles: ", habilidades_candidato_1 | habilidades_candidato_2 | habilidades_requeridas)
```
