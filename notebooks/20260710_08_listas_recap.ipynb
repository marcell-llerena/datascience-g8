```python
lista_de_letras = ["a", "a", "a", "b", "b", "c"]
lista_de_letras.count("b")
```

```python
cont = 0

for letra in lista_de_letras:
    if letra == "a":
        cont += 1

print(cont)
```

### METODOS DE LISTAS

- insert -> inserta un elemento en la posicion o indice dado
- append -> inserta un elemento en la ultima posicion
- clear -> borra el contenido de mi lista
- pop -> elimina el ultimo item de mi lista
- extend -> agrega los elementos de una lista a otra lista
- sort -> ordena los elementos de una lista
- count -> cuenta el valor que se le de en una lista
- len -> sirve para saber cuantos elementos tiene mi lista
