### Operadores aritmeticos

- Suma (+)
- Resta (-)
- Multiplicacion (*)
- Division decimal (/)
- Division entero (//)
- Potencia (**)
- Modulo o resto (%)

```python
numero_1 = 30
numero_2 = 2
print(type(numero_1), type(numero_2))

print("La suma es:", numero_1 + numero_2)
print("La resta es:", numero_1 - numero_2)
print("La multiplicacion es:", numero_1 * numero_2)
print("La division decimal es:", numero_1 / numero_2)
print("La division entera es:", numero_1 // numero_2)
print("La potencia es:", numero_1 ** numero_2)
print("La modulo o resto es:", numero_1 % numero_2)
```

### Operadores logicos

##### Siempre devuelven True o False

- Igual o comparacion (==)
- Diferente (!=)
- Mayor (>)
- Menor (<)
- Mayor o igual (>=)
- Menor o igual (<=)

##### Operadores para concatenar

- y (and)
- o (or)
- no (not)

```python
numero_1 = 30
numero_2 = 2
numero_3 = 30

print(numero_1 == numero_2)
print(numero_1 == numero_3)
```

```python
print(numero_1 != numero_2)
print(numero_1 > numero_2)
print(numero_1 >= numero_3)
```

```python
print(True and True)
```

```python
print(True and False)
```

```python
print(True or False)
```

```python
print(False or False)
```

```python
print(not False)
```

```python
print(not True)
```
