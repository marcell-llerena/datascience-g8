```python
# Encapsulacion

class Cuenta:
    def __init__(self, saldo: float):
        self.__saldo = saldo

    def depositar(self, monto: float):
        self.__saldo += monto

    def retirar(self, monto: float):
        if monto > self.__saldo:
            return "Error saldo insuficiente"
        self.__saldo -= monto

    def mostrar_saldo(self):
        return self.__saldo

cuenta = Cuenta(1000)
print(cuenta.mostrar_saldo())
```

```python
# Herencia
class Animal:
    def __init__(
        self,
        es_domestico: bool,
        num_patas: int,
        nombre: str
    ):
        self.es_domestico = es_domestico
        self.num_patas = num_patas
        self.nombre = nombre

    def comer(self):
        print("Estoy comiendo")

    def beber(self):
        print("Estoy bebiendo agua")

    def es_adoptable(self):
        if self.es_domestico:
            return True
        return False


class Duenio:
    def __init__(self, nombre: str, edad: int):
        self.nombre = nombre
        self.edad = edad

    def adoptar(self):
        return "Estoy adoptando"

# herencia -> es un
# composicion -> tiene un
class Perro(Animal):
    def __init__(
        self,
        es_domestico: bool,
        num_patas: int,
        nombre: str,
        nombre_duenio: str,
        edad_duenio: int
    ):
        super().__init__(es_domestico, num_patas, nombre)
        self.duenio = Duenio(nombre_duenio, edad_duenio)

    def hablar(self):
        return "Gua gua!"


class Gato(Animal):
    def __init__(
        self,
        es_domestico: bool,
        num_patas: int,
        nombre: str
    ):
        super().__init__(es_domestico, num_patas, nombre)

    def hablar(self):
        return "Miau miau!"

    def beber(self):
        print("Yo no bebo agua, solo bebo leche")
```

```python
# Composicion
perro = Perro(num_patas=4, nombre="Fido", es_domestico=True, nombre_duenio="Jorge", edad_duenio=20)
perro.duenio.adoptar()
```

```python
perro = Perro(num_patas=4, nombre="Fido", es_domestico=True)
gato = Gato(True, 4, "Michifus")
```

```python
perro.comer()
```

```python
gato.comer()
```

```python
perro.beber()
```

```python
gato.beber()
```

```python
# Polimorfismo

for animal in [perro, gato]:
    print(animal.hablar())
```
