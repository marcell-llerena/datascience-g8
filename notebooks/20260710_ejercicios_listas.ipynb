```python
# Ejercicio 1
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]
numeros_distinct = []
for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)
print(numeros_distinct)
```

```python
# Ejercicio 2
numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]

numeros_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]

numeros_impares = [
    numero
    for numero in numeros
    if numero % 2 == 1
]

print(f"numeros pares: {numeros_pares}")
print(f"numeros impares: {numeros_impares}")
```

```python
# Ejercicio 3
numeros = [45, 12, 89, 34, 89, 23, 67, 10]
numeros_distinct = []
for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)

numeros_distinct.sort(reverse=True)
print(f"El segundo mas grande es: {numeros_distinct[1]}")
```

```python
# Ejercicio 4
palabras = ['python', 'sol', 'programacion', 'lista', 'computadora', 'if', 'variable']

contador = 0
palabras_con_5_letras = []

for palabra in palabras:
    if len(palabra) > 5:
        contador = contador + 1
        palabras_con_5_letras.append(palabra)

print(f"Existen {contador} palabras con mas de 5 letras")
print(f"Las palabras con mas de 5 letras son: {palabras_con_5_letras}")
```

```python
# Ejercicio 5
notas = [14, 18, 11, 16, 9, 13, 20]

suma_total = 0
total_notas = len(notas)

for nota in notas:
    suma_total = suma_total + nota

promedio = suma_total / total_notas
print(f"El promedio de notas es: {promedio}")

if promedio >= 12:
    print("Aprobo")
else:
    print("Desaprobo")
```

```python
notas = [14, 18, 11, 16, 9, 13, 20]
promedio = round(sum(notas) / len(notas), 2)

print(f"El promedio de notas es: {promedio}")
if promedio >= 12:
    print("Aprobo")
else:
    print("Desaprobo")
```

```python
# Ejercicio 10
colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado']
# colores.reverse()
# print(colores)
ultimo_indice = len(colores) - 1
print(f"ultimo indice: {ultimo_indice}")

colores_invertidos = []
while ultimo_indice >= 0:
    colores_invertidos.append(
        colores[ultimo_indice]
    )
    ultimo_indice = ultimo_indice - 1

print(f"Lista invertida: {colores_invertidos}")
```

```python
# Ejercicio 11
palabras = ['casa', 'elefante', 'sol', 'computadora', 'mesa', 'programacion']
longitudes_palabras = [
    len(palabra)
    for palabra in palabras
]
print(f"Longitudes palabras: {longitudes_palabras}")

lista_listas = []
for item in zip(palabras, longitudes_palabras):
    lista_listas.append(item)

sorted(lista_listas, key=lambda x: x[1])
```

```python
# Ejercicio 16
numeros = [10, 20, 30, 40, 50, 60, 70, 80]
half_list = []
last_list = []
longitud_lista = len(numeros)
indice_mitad = longitud_lista // 2
print(indice_mitad)

for idx, numero in enumerate(numeros):
    if idx < indice_mitad:
        half_list.append(numero)
    else:
        last_list.append(numero)

print(f"Primera mitad: {half_list}")
print(f"Ultima mitad: {last_list}")
```

```python
# Usando slicing
print(f"Primera mitad: {numeros[:indice_mitad]}")
print(f"Ultima mitad: {numeros[indice_mitad :]}")
```

```python
# Ejercicio 20
estudiantes = ['Ana', 'Luis', 'Carlos', 'María', 'Pedro']
notas = [18, 10, 14, 8, 16]

lista_de_aprobados = []
for estudiante, nota in zip(estudiantes, notas):
    if nota > 11:
        lista_de_aprobados.append(estudiante)

print(f"Lista de aprobados: {lista_de_aprobados}")
```
