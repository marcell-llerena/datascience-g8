```python
# Ejercicio 1
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]
numeros_distinct = []
for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)
print(numeros_distinct)
```

```python
# Ejercicio 2
numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]

numeros_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]

numeros_impares = [
    numero
    for numero in numeros
    if numero % 2 == 1
]

print(f"numeros pares: {numeros_pares}")
print(f"numeros impares: {numeros_impares}")
```

```python
# Ejercicio 3
numeros = [45, 12, 89, 34, 89, 23, 67, 10]
numeros_distinct = []
for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)

numeros_distinct.sort(reverse=True)
print(f"El segundo mas grande es: {numeros_distinct[1]}")
```

```python
# Ejercicio 4
palabras = ['python', 'sol', 'programacion', 'lista', 'computadora', 'if', 'variable']

contador = 0
palabras_con_5_letras = []

for palabra in palabras:
    if len(palabra) > 5:
        contador = contador + 1
        palabras_con_5_letras.append(palabra)

print(f"Existen {contador} palabras con mas de 5 letras")
print(f"Las palabras con mas de 5 letras son: {palabras_con_5_letras}")
```
