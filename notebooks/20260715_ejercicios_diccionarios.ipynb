```python
# Ejercicio 1

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3}

inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5}

inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}
```

```python
inventario_total = dict(inventario_norte)
print(inventario_total)
```

```python
for llave, valor in inventario_sur.items():
    print(f"Llave: {llave} valor: {valor}")
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor
    print(f"inventario_total - llave: {inventario_total[llave]}")
    print('='* 10)
```

```python
print(inventario_total)
```

```python

for llave, valor in inventario_centro.items():
    print(f"llave: {llave} | valor: {valor}")
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor
```

```python
print(inventario_total)
```

```python
llaves_ordenadas = sorted(inventario_total.keys())
print(llaves_ordenadas)
```

```python
inventario_total_con_llaves_ordenadas = {}

for llave in llaves_ordenadas:
    inventario_total_con_llaves_ordenadas[llave] = inventario_total[llave]

print(inventario_total)
print(inventario_total_con_llaves_ordenadas)
```

```python
# Ejercicio 2
calificaciones = {
    "Ana": [17, 18, 16, 19],
    "Luis": [15, 16, 17, 16],
    "Carla": [18, 17, 19, 18],
    "Pedro": [16, 16, 16, 16],
    "Lucía": [15, 17, 16, 16]
}

estadisticas = {}

for estudiante, notas in calificaciones.items():
    print(estudiante, notas)
    estadisticas[estudiante] = {
        "minima_nota": min(notas),
        "maxima_nota": max(notas),
        "promedio": sum(notas) / len(notas)
    }
    print(estadisticas[estudiante])
    print('='*20)
```

```python
for estudiante, stats in estadisticas.items():
    print(estudiante)
    print(stats)
    print('='*20)
```

```python
ranking = []
for nombre, stats in estadisticas.items():
    ranking.append(
        (nombre, stats["promedio"])
    )
print(ranking)

# ranking = sorted(ranking, key=lambda x: x[1], reverse=True)
# print(ranking)
```

```python
ranking = sorted(ranking, key=lambda item: item[1], reverse=True)
print(ranking)
```

```python
ranking[2:] = sorted(ranking[2:], key=lambda item: item[0], reverse=False)
print(ranking)
```

```python
sorted(ranking, key=lambda x: (x[1], x[0]), reverse=True)
```

```python
# Ejercicios 3
usuarios = {
    " ANA_01 ": {"nombre": " ana torres ", "correo": " ANA@MAIL.COM ", "edad": "24"},
    "LUIS_02": {"nombre": "luis perez", "correo": "Luis@Mail.com", "edad": "treinta"},
    " carla_03": {"nombre": "CARLA RUIZ", "correo": "carla@mail.com", "edad": "27"}
}

usuarios_limpios = {}

for usuario, datos in usuarios.items():
    print(usuario, datos)

    llave_limpia = usuario.strip()
    correo_minuscula = datos['correo'].lower().strip()
    nombre_limpio = datos['nombre'].strip().upper()

    try:
        edad_entera = int(datos['edad'])
    except:
        edad_entera = None

    datos_limpios = {
        'nombre': nombre_limpio,
        'correo': correo_minuscula,
        'edad': edad_entera
    }

    usuarios_limpios[llave_limpia] = datos_limpios

    print('=' * 20)
```

```python
# Ejercicio 4
texto = "Python es poderoso; Python es flexible, y aprender Python puede ser muy divertido."
palabras_vacias = {"es", "y", "ser", "muy", "puede"}

contador = {}

for palabra in texto.split(" "):
    palabra_limpia = palabra.lower().replace(";", "").replace(",", "").replace(".", "")

    if palabra_limpia in palabras_vacias:
        continue

    # if palabra_limpia not in contador:
    #     contador[palabra_limpia] = 0
    # contador[palabra_limpia] += 1

    contador.setdefault(palabra_limpia, 0)
    contador[palabra_limpia] += 1

print(contador)
```
