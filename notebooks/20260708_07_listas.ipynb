### LISTA

- append -> inserta un elemento al final de la lista
- insert -> inserta un elemento en un posicion dada
- remove -> remueve un elemento de la lista
- pop -> remueve el ultimo elemento
- sort -> ordenaba la lista
- len -> longitud o contar cuantos elementos tiene la lista

```python
list(range(6))
```
