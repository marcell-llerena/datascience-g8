### Condicionales

SI se cumple esta CONDICION:
    - Proceso 1
SI se cumple esta otra CONDICION:
    - Proceso 2
Si no se cumple ninguna de las otras 2 condiciones:
    - Proceso por defecto o por default


si soy mayor de 18 anios:
   - soy mayor de edad
si edad es mayor a 13 anios:
   - soy adolescente
si no soy mayor de 13 anios:
   - soy un infante








```python
edad = 30

if edad >= 18:
    print("soy un adulto")
elif edad >= 13:
    print("soy una adolescente")
else:
    print("soy un infante")
```
